# Save nested JSON to a Pandas Dataframe

```python
import pandas as pd
import json
from pprint import pprint as pp
```

```python
with open('brian_sample.json') as brian_in:
    raw_data = json.load(brian_in)
pp(raw_data)
```

```python
fields = list(raw_data.keys())
fields.remove('traces')

df_inputs = pd.json_normalize(
    raw_data,
    record_path=["traces", "inputs"],
    meta = fields,
)
df_inputs['side'] = ["inputs"] * len(df_inputs)

df_outputs = pd.json_normalize(
    raw_data,
    record_path=["traces", "outputs"],
    meta = fields,
)
df_outputs['side'] = ["outputs"] * len(df_outputs)

df = pd.concat([df_inputs, df_outputs])
df
```
